#### **2. HOMOGENEIZACIÓN EN EL NOMBRADO DE COLUMNAS**

- En esta sección se estandarizan los nombres de las columnas para garantizar consistencia y facilitar su uso en las fases posteriores del análisis.
- Primero, se normalizan los nombres al formato `snake_case`, eliminando espacios y caracteres especiales.
- Posteriormente, se renombran aquellas variables con nomenclatura ambigua para mejorar su claridad semántica (por ejemplo, `Date&Time` se renombra a `date_time`).
- Este proceso mejora la legibilidad del dataset y facilita su manipulación en el flujo de limpieza y análisis.

```python
def normalizar_nombres_columnas(lista_columnas, mostrar_resumen=True):
    """
    Normaliza los nombres de columnas a formato snake_case estándar.

    Transformaciones aplicadas:
    - Eliminación de espacios al inicio y final
    - Conversión de CamelCase y PascalCase a snake_case
    - Sustitución de caracteres no alfanuméricos por guión bajo (_)
    - Conversión a minúsculas
    - Eliminación de guiones bajos duplicados
    - Eliminación de guiones bajos al inicio y final

    Parámetros
    ----------
    lista_columnas : list of str
        Lista de nombres de columnas originales.

    mostrar_resumen : bool, default=True
        Si True, muestra un resumen de la transformación realizada.

    Returns
    -------
    list of str
        Lista de nombres normalizados en formato snake_case.

    Ejemplo
    -------
    >>> normalizar_nombres_columnas(["Date&Time", "UserID", "App Name"])
    ['date_time', 'user_id', 'app_name']
    """

    nombres_normalizados = []
    contador = {}

    for nombre in lista_columnas:

        original = nombre

        # quitar espacios
        limpia = nombre.strip()

        # eliminar acentos
        limpia = unicodedata.normalize("NFKD", limpia)
        limpia = limpia.encode("ascii", "ignore").decode("utf-8")

        # camelCase → snake_case
        limpia = re.sub(r'([a-z0-9])([A-Z])', r'\1_\2', limpia)

        # caracteres no alfanuméricos → _
        limpia = re.sub(r'[^0-9a-zA-Z]+', '_', limpia)

        # minúsculas
        limpia = limpia.lower()

        # quitar duplicados de _
        limpia = re.sub(r'_+', '_', limpia)

        # quitar _ extremos
        limpia = limpia.strip('_')

        # 🔹 asegurar unicidad
        if limpia in contador:
            contador[limpia] += 1
            limpia = f"{limpia}_{contador[limpia]}"
        else:
            contador[limpia] = 0

        nombres_normalizados.append(limpia)

    if mostrar_resumen:
        print("Normalización de nombres de columnas completada.")
        print(f"Columnas procesadas: {len(lista_columnas)}")
        print("Transformaciones aplicadas:")
        for original, nuevo in zip(lista_columnas, nombres_normalizados):
            if original != nuevo:
                print(f"  {original} → {nuevo}")

    return nombres_normalizados
```

#### **3. ELIMINACIÓN DE COLUMNAS NO INFORMATIVAS**

- En esta sección se identifican y eliminan variables que no aportan información relevante para el análisis y que pueden introducir ruido en el dataset.
- Primero, se eliminan aquellas columnas completamente nulas (100% valores faltantes), ya que no contienen información utilizable: `eliminar_columnas_vacias`.
- Posteriormente, se eliminan las columnas constantes, es decir, aquellas que presentan un único valor en todas las filas y no aportan variabilidad ni capacidad explicativa `eliminar_columnas_constantes`.
- Por último, se eliminan aquellas columnas no consideras de interés para el análisis de datos: `eliminar_columnas_sin_interes`.
- Este proceso permite reducir la dimensionalidad del dataset y simplificar las fases posteriores de exploración, análisis y visualización.

```python
def eliminar_columnas_vacias(df, mostrar_resumen=True):
    """
    Elimina columnas que contienen únicamente valores nulos.

    Parámetros
    ----------
    df : pandas.DataFrame
        DataFrame a procesar.

    mostrar_resumen : bool, default=True
        Si True, muestra qué columnas se han eliminado.

    Returns
    -------
    pandas.DataFrame
        DataFrame sin columnas completamente nulas.
    """

    columnas_vacias = df.columns[df.isnull().all()]

    if mostrar_resumen and len(columnas_vacias) > 0:
        print(f"Columnas eliminadas ({len(columnas_vacias)}):")
        for col in columnas_vacias:
            print(f"  - {col}")

    df = df.drop(columns=columnas_vacias)

    return df
```

```python
for i, df in enumerate(dfs):
    print(f"\n===== DF{i+1} =====")
    dfs[i] = eliminar_columnas_vacias(df)
```

```python
def eliminar_columnas_constantes(df, excepciones=None, mostrar_resumen=True):
    """
    Elimina columnas que contienen un único valor único en todo el DataFrame.

    Este tipo de columnas no aporta información analítica, ya que no presentan
    variabilidad entre observaciones. Pueden aparecer frecuentemente en datasets
    estadísticos o administrativos como metadatos constantes.

    Parámetros
    ----------
    df : pandas.DataFrame
        DataFrame a procesar.

    excepciones : list, default=None
        Lista de columnas que no deben eliminarse aunque tengan un único valor.
        Útil para mantener variables estructurales como 'unit'.

    mostrar_resumen : bool, default=True
        Si True, muestra por consola las columnas que han sido eliminadas.

    Returns
    -------
    pandas.DataFrame
        DataFrame sin columnas constantes (excepto las indicadas en excepciones).

    Ejemplo
    -------
    >>> eliminar_columnas_constantes(df, excepciones=["unit"])
    Columnas constantes eliminadas (2):
      - dataset
      - freq
    """
    
    if excepciones is None:
        excepciones = []

    columnas_constantes = [
        col for col in df.columns
        if df[col].nunique(dropna=False) == 1 and col not in excepciones
    ]

    if mostrar_resumen and len(columnas_constantes) > 0:
        print(f"Columnas constantes eliminadas ({len(columnas_constantes)}):")
        for col in columnas_constantes:
            print(f"  - {col}")

    df = df.drop(columns=columnas_constantes)

    return df
```

```python
def eliminar_columnas_sin_interes(df, mostrar_resumen=True):
    """
    Elimina columnas que no se consideran relevantes para el análisis.

    Incluye:
    - Flags y metadatos típicos (calidad/estado/confidencialidad).
    - Dimensiones/códigos que no se van a explotar en este análisis (ocupación, educación, sector, unidad).

    Nota: esta función asume que los nombres de columnas ya están normalizados
    (snake_case) en pasos previos del pipeline.

    Parameters
    ----------
    df : pandas.DataFrame
        DataFrame de entrada.
    mostrar_resumen : bool, default=True
        Si True, imprime qué columnas se eliminan.

    Returns
    -------
    pandas.DataFrame
        DataFrame sin las columnas eliminadas.
    """

    columnas_a_eliminar = [
        # --- FLAGS / METADATA ---
        "obs_flag",
        "observation_status_flag_v2_structure",
        "conf_status",
        "confidentiality_status_flag",

        # --- DIMENSIONES (NO EN ALCANCE) ---
        "isco08",
        "isced11",
        "nace_r2",
        "unit",
    ]

    cols_drop = [c for c in columnas_a_eliminar if c in df.columns]

    if mostrar_resumen:
        print(f"Columnas eliminadas ({len(cols_drop)}): {cols_drop}")

    return df.drop(columns=cols_drop, errors="ignore")
```

#### **4. TRATAMIENTO DE VALORES NULOS**

- En esta sección se gestionan los valores nulos del dataset para garantizar la consistencia de la información antes de iniciar el análisis exploratorio.
- En primer lugar, se eliminan las filas donde la variable principal de análisis (`obs_value`) presenta valores nulos, ya que estas observaciones no aportan información útil para el estudio del indicador.
- En el resto de variables categóricas, los valores nulos se sustituyen por la categoría `Unknown`, permitiendo mantener las observaciones sin introducir sesgos en el análisis.
- Los nulos en las variables numéricas, si los hubiera, se mantienen.
- Este procedimiento asegura que las métricas analizadas se basen únicamente en datos válidos, al mismo tiempo que se preserva la mayor cantidad posible de información en las variables descriptivas.

```python
def tratamiento_nulos(df, col_valor="obs_value", fill_unknown="Unknown", mostrar_resumen=True):
    """
    Trata los nulos con una regla dirigida a datasets tipo Eurostat/indicadores:

    Reglas
    ------
    1) Elimina filas donde la columna principal de análisis (col_valor) sea nula.
       (Ej: 'obs_value' suele ser el valor del indicador).
    2) En el resto de columnas categóricas (object/category/string), rellena nulos con 'Unknown'.
    3) No modifica nulos en columnas numéricas distintas de col_valor (se conservan tal cual).

    Parámetros
    ----------
    df : pandas.DataFrame
        DataFrame de entrada.
    col_valor : str, default="obs_value"
        Nombre de la columna cuyo nulo implica eliminación de la fila.
    fill_unknown : str, default="Unknown"
        Valor con el que se rellenan los nulos en columnas categóricas.
    mostrar_resumen : bool, default=True
        Si True, imprime un resumen de filas eliminadas y nulos rellenados.

    Returns
    -------
    pandas.DataFrame
        DataFrame tras:
        - eliminar filas con nulos en col_valor
        - rellenar nulos en categóricas con fill_unknown

    Nota
    ----
    Para actualizar el dataframe:
    >>> df = tratamiento_nulos(df, col_valor="obs_value")
    """

    filas_iniciales = df.shape[0]

    # eliminar filas con nulos en columna valor SOLO si existe
    if col_valor in df.columns:
        nulos_valor = df[col_valor].isna().sum()
        df = df.dropna(subset=[col_valor]).copy()
        filas_eliminadas = filas_iniciales - df.shape[0]
    else:
        filas_eliminadas = 0
        if mostrar_resumen:
            print(f"La columna '{col_valor}' no existe en este dataset")

    # rellenar categóricas
    cols_cat = df.select_dtypes(include=["object","category"]).columns
    df[cols_cat] = df[cols_cat].fillna("Unknown")

    if mostrar_resumen:
        print("Tratamiento de nulos completado")
        print(f"Filas iniciales: {filas_iniciales}")
        print(f"Filas eliminadas por nulos en '{col_valor}': {filas_eliminadas}")
        print(f"Filas finales: {df.shape[0]}")

    return df
```

#### **5. PREPARACIÓN DEL DATASET DE SKILLS**

- En esta sección se prepara el dataset de ofertas de empleo para analizar las habilidades demandadas en el mercado laboral tecnológico.
- Primero, se crea una copia del dataset original y se seleccionan únicamente las variables relevantes para el análisis (`job_title`, `seniority_level`, `status`, `industry`, `company_size` y `skills`).
- Posteriormente, se eliminan las observaciones donde la lista de habilidades está vacía, ya que no aportan información sobre competencias requeridas.
- Finalmente, se transforma la columna `skills` para que cada habilidad aparezca en una fila independiente, generando así un dataset estructurado que facilita el análisis de frecuencia y distribución de habilidades.

```python
def preparar_skills_dataset(df, fill_unknown="Unknown"):
    """
    Prepara un dataset de ofertas de empleo para analizar skills.

    Procedimiento
    -------------
    1. Trabaja sobre una copia del DataFrame original.
    2. Elimina filas donde la columna 'skills' esté vacía o sea nula.
    3. Conserva únicamente las columnas relevantes:
       job_title, seniority_level, status, industry, company_size y skills.
    4. Normaliza la columna 'company_size':
       - elimina separadores de miles
       - convierte a numérico cuando es posible
       - valores no convertibles se sustituyen por 'Unknown'.
    5. Rellena valores nulos en columnas categóricas con 'Unknown'.
    6. Genera un nuevo DataFrame donde cada skill ocupa una fila
       utilizando explode().

    Parameters
    ----------
    df : pandas.DataFrame
        DataFrame de entrada con las ofertas de empleo.
    fill_unknown : str, default="Unknown"
        Valor utilizado para rellenar nulos en columnas categóricas
        o valores no interpretables.

    Returns
    -------
    pandas.DataFrame
        Nuevo DataFrame donde cada fila representa una skill individual.
    """

    # copiar dataframe
    df2 = df.copy()

    # crear ID por oferta
    df2["offer_id"] = (
        df2[
            [
                "job_title",
                "seniority_level",
                "status",
                "industry",
                "company_size"
            ]
        ]
        .astype(str)
        .agg("_".join, axis=1)
        .factorize()[0] + 1
    )

    # eliminar filas sin skills
    df2 = df2[df2["skills"].notna()]
    df2 = df2[df2["skills"].astype(str).str.strip().ne("")]
    df2 = df2[df2["skills"].astype(str).str.strip().ne("[]")]

    # seleccionar columnas relevantes
    columnas = [
        "offer_id",
        "job_title",
        "seniority_level",
        "status",
        "industry",
        "company_size",
        "skills"
    ]

    df2 = df2[columnas].copy()

    # convertir skills a lista
    df2["skills"] = df2["skills"].astype(str).apply(ast.literal_eval)

    # normalizar company_size
    df2["company_size"] = pd.to_numeric(
        df2["company_size"].astype(str).str.replace(",", "", regex=False),
        errors="coerce"
    ).fillna(fill_unknown)

    # rellenar nulos en categóricas
    cols_cat = df2.select_dtypes(include=["object"]).columns
    df2[cols_cat] = df2[cols_cat].fillna(fill_unknown)

    # una skill por fila
    df_skills = df2.explode("skills").copy()
    df_skills["skills"] = df_skills["skills"].astype(str).str.lower().str.strip()

    # eliminar skills vacías o nulas
    df_skills = df_skills[(df_skills["skills"] != "") & (df_skills["skills"] != "nan")]

    return df_skills
```

#### **6. LIMPIEZA DE VARIABLES CATEGÓRICAS**

- En esta sección se estandariza el formato de las variables categóricas.
- Se eliminan espacios al inicio y al final de los valores y se aplica un formato homogéneo de capitalización.
- Este proceso garantiza la consistencia de las categorías y evita la duplicación de valores equivalentes con distinta representación.

```python
def limpiar_categoricas(df, columnas=None):
    """
    Limpia variables categóricas eliminando espacios y unificando el formato del texto.

    Aplica .str.strip().str.capitalize() para garantizar consistencia en las categorías.

    Parámetros
    ----------
    df : pandas.DataFrame
        DataFrame a procesar.

    columnas : list, default=None
        Columnas a limpiar. Si es None, se aplicará a todas las columnas categóricas.

    Devuelve
    --------
    pandas.DataFrame
        DataFrame con las variables categóricas limpias.
    """

    # Si no se especifican columnas, detectar todas las categóricas
    if columnas is None:
        columnas = df.select_dtypes(include="O").columns

    # Aplicar limpieza básica
    for col in columnas:
        df[col] = df[col].str.strip().str.upper()

    return df
```

#### **7. AGRUPACIÓN Y SIMPLIFICACIÓN DE VARIABLES CATEGÓRICAS**

- En esta sección se simplifican y agrupan algunas variables categóricas para facilitar su interpretación en el análisis posterior.
- Primero, se renombran columnas con nombres excesivamente largos por etiquetas más intuitivas (`occupation`, `education`, `sector`).
- Posteriormente, se generan nuevas variables agregadas que reducen el número de categorías, como `company_size_group`, `education_group` o `sector_group`.
- Este proceso permite trabajar con dimensiones analíticas más claras y comparables entre los diferentes datasets.

```python
def aplicar_agrupaciones(df):
    """
    Aplica simplificaciones y agrupaciones a variables categóricas del dataset.

    Transformaciones realizadas:
    - Renombrado de columnas largas a nombres más intuitivos (`occupation`, `education`, `sector`).
    - Agrupación del tamaño de empresa en categorías (`company_size_group`).
    - Simplificación de textos en ocupación y nivel educativo.
    - Creación de grupos agregados de sector económico (`sector_group`).

    Devuelve
    --------
    pandas.DataFrame
        DataFrame con las nuevas variables derivadas para facilitar el análisis.
    """

    cols_iniciales = set(df.columns)

    # renombrar columnas largas a nombres más intuitivos (si existen)
    df = df.rename(columns={
        "international_standard_classification_of_occupations_2008_isco_08": "occupation",
        "international_standard_classification_of_education_isced_2011": "education",
        "statistical_classification_of_economic_activities_in_the_european_community_nace_rev_2": "sector",
    })

    # company size (numérico) -> grupo
    if "company_size" in df.columns:
        size = pd.to_numeric(df["company_size"], errors="coerce")
        df["company_size_group"] = pd.cut(
            size, [0, 50, 250, 1000, 5000, 1e9],
            labels=["STARTUP", "SME", "MID_SIZE", "LARGE", "ENTERPRISE"]
        )

    # occupation -> texto corto
    if "occupation" in df.columns:
        s = df["occupation"].astype(str)
        df["occupation_short"] = s

        df.loc[s.str.contains("SERVICE MANAGERS", na=False), "occupation_short"] = "ICT SERVICE MANAGERS"
        df.loc[s.str.contains("SOFTWARE AND APPLICATIONS DEVELOPERS", na=False), "occupation_short"] = "SOFTWARE & APP DEVELOPERS"
        df.loc[s.str.contains("DATABASE AND NETWORK PROFESSIONALS", na=False), "occupation_short"] = "DATABASE & NETWORK PROFESSIONALS"
        df.loc[s.str.contains("OPERATIONS AND USER SUPPORT TECHNICIANS", na=False), "occupation_short"] = "ICT OPS & SUPPORT TECHNICIANS"
        df.loc[s.str.contains("TELECOMMUNICATIONS AND BROADCASTING TECHNICIANS", na=False), "occupation_short"] = "TELECOM & BROADCAST TECHNICIANS"
        df.loc[s.str.contains("INSTALLERS AND REPAIRERS", na=False), "occupation_short"] = "ELECTRONICS & TELECOM INSTALLERS"
        df.loc[s.str.contains("OTHER INFORMATION AND COMMUNICATIONS TECHNOLOGY SPECIALISTS", na=False), "occupation_short"] = "OTHER ICT SPECIALISTS"

    # education -> texto corto + education_group
    if "education" in df.columns:
        s = df["education"].astype(str)
        df["education_short"] = s

        df.loc[s.str.contains("LESS THAN PRIMARY", na=False), "education_short"] = "ISCED 0_4 NON_TERTIARY"
        df.loc[s.str.contains("TERTIARY EDUCATION", na=False), "education_short"] = "ISCED 5_8 TERTIARY"
        df.loc[s.str.contains("NO RESPONSE", na=False), "education_short"] = "UNKNOWN"

        df["education_group"] = "UNKNOWN"
        df.loc[s.str.contains("TERTIARY", na=False), "education_group"] = "HIGHER_EDUCATION"
        df.loc[s.str.contains("LESS THAN PRIMARY", na=False), "education_group"] = "LOW_MEDIUM_EDUCATION"

    elif "isced11" in df.columns:
        df["education_group"] = df["isced11"].replace({
            "ED0-4": "LOW_MEDIUM_EDUCATION",
            "ED5-8": "HIGHER_EDUCATION",
            "NRP": "UNKNOWN"
        })

    # sector -> macrosector
    if "sector" in df.columns:
        s = df["sector"].astype(str)

        df["sector_group"] = "OTHER"
        df.loc[s.str.contains("AGRICULTURE|FORESTRY|FISHING|MINING|QUARRYING", na=False), "sector_group"] = "PRIMARY"
        df.loc[s.str.contains("MANUFACTUR|ELECTRICITY|GAS|STEAM|WATER SUPPLY|SEWERAGE|WASTE", na=False), "sector_group"] = "INDUSTRY"
        df.loc[s.str.contains("CONSTRUCTION", na=False), "sector_group"] = "CONSTRUCTION"
        df.loc[s.str.contains("WHOLESALE|RETAIL|TRANSPORTATION|STORAGE|ACCOMMODATION|FOOD SERVICE", na=False), "sector_group"] = "TRADE_TRANSPORT_HOSPITALITY"
        df.loc[s.str.contains("INFORMATION AND COMMUNICATION|TELECOMMUNICATIONS|COMPUTER PROGRAMMING|ICT", na=False), "sector_group"] = "ICT"
        df.loc[s.str.contains("FINANCIAL|INSURANCE", na=False), "sector_group"] = "FINANCE"
        df.loc[s.str.contains("REAL ESTATE", na=False), "sector_group"] = "REAL_ESTATE"
        df.loc[s.str.contains("PROFESSIONAL|SCIENTIFIC|TECHNICAL|LEGAL|ACCOUNTING|CONSULTANCY|RESEARCH|ADVERTISING", na=False), "sector_group"] = "PROFESSIONAL_SERVICES"
        df.loc[s.str.contains("ADMINISTRATIVE|SUPPORT SERVICE|RENTAL|LEASING|EMPLOYMENT|SECURITY", na=False), "sector_group"] = "ADMIN_SUPPORT"
        df.loc[s.str.contains("PUBLIC ADMINISTRATION|DEFENCE|EDUCATION|HUMAN HEALTH|SOCIAL WORK", na=False), "sector_group"] = "PUBLIC_SERVICES"
        df.loc[s.str.contains("NO RESPONSE", na=False), "sector_group"] = "UNKNOWN"

    # columnas nuevas creadas
    cols_finales = set(df.columns)
    cols_nuevas = cols_finales - cols_iniciales

    if cols_nuevas:
        print("Columnas creadas:", sorted(cols_nuevas))
    else:
        print("No se han creado nuevas columnas")

    return df
```

#### **8. LIMPIEZA Y TRANSFORMACIÓN DEL DATASET DE SALARIOS**

- En esta sección se aplican transformaciones específicas sobre el dataset de salarios para estandarizar variables categóricas y facilitar su análisis posterior.
- Primero, se normalizan distintas variables descriptivas mediante mapeos simplificados, como `experience_level`, `employment_type`, `company_size` o los países de `company_location` y `employee_residence`.
- Posteriormente, se generan nuevas variables derivadas relevantes para el análisis, como `remote_type`, `role_category`, `salary_band` o `international_job`.
- Finalmente, se eliminan duplicados y variables intermedias que ya no son necesarias, dejando el dataset preparado para la fase de análisis exploratorio.

```python
def limpiar_dataset_salarios(df):
    """
    Aplica la limpieza y transformación específica del dataset de salarios.

    Pasos
    -----
    1. Imputa valores nulos en remote_ratio.
    2. Estandariza experience_level, employment_type y company_size.
    3. Convierte company_location y employee_residence a códigos país.
    4. Crea la variable remote_type.
    5. Limpia espacios en job_title.
    6. Elimina la columna remote_ratio.
    7. Elimina duplicados según variables clave.
    8. Crea role_category a partir de job_title.
    9. Genera salary_band por cuartiles.
    10. Crea la variable international_job.

    Returns
    -------
    pandas.DataFrame
        DataFrame limpio y transformado.
    """

    df = df.copy()

    # imputar remote_ratio
    df["remote_ratio"] = df["remote_ratio"].fillna("Unknown")

    # mapear experience level
    experience_map = {
        "Entry": "EN",
        "Junior": "EN",
        "Mid": "MI",
        "Senior": "SE",
        "Expert": "EX"
    }
    df["experience_level"] = df["experience_level"].replace(experience_map)

    # mapear employment type
    employment_map = {
        "Full-Time": "FT",
        "Part-Time": "PT",
        "Contract": "CT",
        "Freelance": "FL"
    }
    df["employment_type"] = df["employment_type"].replace(employment_map)

    # mapear company size
    company_size_map = {
        "Small": "S",
        "Medium": "M",
        "Large": "L"
    }
    df["company_size"] = df["company_size"].replace(company_size_map)

    # mapear company_location y employee_residence
    country_map = {
        "United States": "US",
        "United Kingdom": "GB",
        "Germany": "DE",
        "Sweden": "SE",
        "India": "IN",
        "Korea, Republic of": "KR",
        "Spain": "ES",
        "Estonia": "EE",
        "France": "FR",
        "Philippines": "PH",
        "Canada": "CA",
        "Australia": "AU",
        "Turkey": "TR",
        "Portugal": "PT",
        "Brazil": "BR",
        "Switzerland": "CH",
        "Andorra": "AD",
        "Netherlands": "NL",
        "Ecuador": "EC",
        "Mexico": "MX",
        "Israel": "IL",
        "Nigeria": "NG",
        "Saudi Arabia": "SA",
        "Colombia": "CO",
        "Poland": "PL",
        "Norway": "NO",
        "Ghana": "GH",
        "Argentina": "AR",
        "Japan": "JP",
        "Russian Federation": "RU",
        "South Africa": "ZA",
        "Italy": "IT",
        "Hong Kong": "HK",
        "Central African Republic": "CF",
        "Finland": "FI",
        "Ukraine": "UA",
        "Ireland": "IE",
        "Singapore": "SG",
        "Slovenia": "SI",
        "Thailand": "TH",
        "Croatia": "HR",
        "Armenia": "AM",
        "Bosnia and Herzegovina": "BA",
        "Kenya": "KE",
        "Latvia": "LV",
        "Romania": "RO",
        "Pakistan": "PK",
        "Lithuania": "LT",
        "Iran, Islamic Republic of": "IR",
        "Bahamas": "BS",
        "Hungary": "HU",
        "Austria": "AT",
        "Puerto Rico": "PR",
        "American Samoa": "AS",
        "Greece": "GR",
        "Denmark": "DK",
        "Belgium": "BE",
        "Indonesia": "ID",
        "Egypt": "EG",
        "United Arab Emirates": "AE",
        "Malaysia": "MY",
        "Honduras": "HN",
        "Czechia": "CZ",
        "Algeria": "DZ",
        "Iraq": "IQ",
        "China": "CN",
        "New Zealand": "NZ",
        "Chile": "CL",
        "Moldova, Republic of": "MD",
        "Luxembourg": "LU",
        "Malta": "MT",
        "Peru": "PE",
        "Uzbekistan": "UZ",
        "Georgia": "GE",
        "Kuwait": "KW",
        "Cyprus": "CY",
        "Costa Rica": "CR",
        "Bolivia, Plurinational State of": "BO",
        "Dominican Republic": "DO",
        "Tunisia": "TN",
        "Viet Nam": "VN",
        "Serbia": "RS",
        "Bulgaria": "BG",
        "Jersey": "JE",
        "Uganda": "UG"
    }

    df["company_location"] = df["company_location"].replace(country_map)
    df["employee_residence"] = df["employee_residence"].replace(country_map)

# mapear currency

    currency_map = {
        "United States Dollar": "USD",
        "Euro": "EUR",
        "British Pound Sterling": "GBP",
        "Indian Rupee": "INR",
        "Canadian Dollar": "CAD",
        "Australian Dollar": "AUD",
        "Polish Zloty": "PLN",
        "Singapore Dollar": "SGD",
        "Swiss Franc": "CHF",
        "Japanese Yen": "JPY",
        "Hungarian Forint": "HUF",
        "Turkish Lira": "TRY",
        "Danish Krone": "DKK",
        "Brazilian Real": "BRL",
        "Thai Baht": "THB",
        "Norwegian Krone": "NOK",
        "Philippine Peso": "PHP",
        "South African Rand": "ZAR",
        "Israeli New Shekel": "ILS",
        "Hong Kong Dollar": "HKD",
        "Mexican Peso": "MXN",
        "Chilean Peso": "CLP"
    }

    df["salary_currency"] = df["salary_currency"].replace(currency_map)

    # mapear remote ratio
    remote_map = {
        0: "On-site",
        50: "Hybrid",
        100: "Remote",
        "Unknown": "Unknown"
    }
    df["remote_type"] = df["remote_ratio"].map(remote_map)

    # limpiar espacios en job_title
    df["job_title"] = df["job_title"].str.strip()

    # eliminar remote_ratio
    df = df.drop(columns="remote_ratio")

    # eliminar duplicados
    df = df.drop_duplicates(
        subset=[
            "work_year",
            "job_title",
            "salary_in_usd",
            "employee_residence",
            "company_location"
        ]
    )

    # categorizar rol
    def categorize_role(title):
        title = title.lower()

        if "research" in title:
            return "Research"
        elif "analyst" in title or "analytics" in title or "bi" in title or "business intelligence" in title:
            return "Data Analyst / BI"
        elif "machine learning" in title or "ml" in title or "ai" in title or "deep learning" in title or "computer vision" in title or "nlp" in title:
            return "ML / AI Engineer"
        elif "scientist" in title or "data science" in title:
            return "Data Scientist"
        elif "engineer" in title:
            return "Data Engineer"
        elif "architect" in title or "etl" in title or "integration" in title or "modeler" in title or "modeller" in title or "developer" in title:
            return "Data Architecture / Infrastructure"
        elif "manager" in title or "head" in title or "director" in title or "lead" in title or "strategist" in title or "product owner" in title:
            return "Management"
        elif "specialist" in title and "data" in title:
            return "Data Analyst / BI"
        elif "operations" in title:
            return "Data Engineer"
        elif "management" in title:
            return "Data Architecture / Infrastructure"
        elif "autonomous" in title:
            return "ML / AI Engineer"
        else:
            return "Other"

    df["role_category"] = df["job_title"].apply(categorize_role)

    # bandas salariales
    df["salary_band"] = pd.qcut(
        df["salary_in_usd"],
        q=4,
        labels=["Low", "Medium", "High", "Very High"]
    )

    # empleo internacional
    df["international_job"] = df["company_location"] != df["employee_residence"]

    return df
```
